```python


snc.jointplot(x='rainfall_mean', y='PM2.5_mean', data=df2, kind='scatter')
plt.show()
```

```python
plt.hexbin(df2['rainfall_mean'], df2['PM2.5_mean'], gridsize=20, cmap='Blues')
plt.xlabel('Rainfall Mean')
plt.ylabel('PM2.5 Mean')
plt.title('Relationship between PM2.5 Mean and Rainfall Mean')
plt.colorbar(label='Count')
plt.show()
```

```python
snc.lineplot(x='month', y='PM2.5_mean', data=df2, ci='sd')
plt.xlabel('Month')
plt.ylabel('PM2.5 Mean')
plt.title('Trend of PM2.5 Mean Over Time')
plt.show()
```

```python
snc.pairplot(df2, vars=['rainfall_mean', 'PM2.5_mean'], hue='month', palette='viridis')
plt.suptitle('Pair Plot of Rainfall Mean and PM2.5 Mean by Month')
plt.show()
```

```python
# Correlation Heatmap:
# Visualize the correlation between different variables, including 'rainfall_mean' and 'PM2.5_mean', using a heatmap.
# python
# A positive or negative correlation between 'rainfall_mean' and 'PM2.5_mean' can give insights into their relationship.
correlation_matrix = df2[['rainfall_mean', 'PM2.5_mean']].corr()
snc.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python


# Calculate the correlation matrix including 'month'
correlation_matrix = df2[['rainfall_mean', 'PM2.5_mean', 'month']].corr()

# Create a heatmap with annotations
snc.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap - Month Wise')
plt.show()
```

```python

# Calculate the correlation matrix for each month
correlation_matrix_monthly = df2.groupby('month')[['rainfall_mean', 'PM2.5_mean']].corr().unstack()['rainfall_mean']

# Create a heatmap with annotations
snc.heatmap(correlation_matrix_monthly, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap: Rainfall Mean vs. PM2.5 Mean (Month-wise)')
plt.xlabel('Variable')
plt.ylabel('Month')
plt.show()
```

```python
snc.regplot(x='rainfall_mean', y='PM2.5_mean', data=df2, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.xlabel('Rainfall Mean')
plt.ylabel('PM2.5 Mean')
plt.title('Scatter Plot with Regression Line')
plt.show()
```

```python
snc.boxplot(x=pd.cut(df2['rainfall_mean'], bins=5), y='PM2.5_mean', data=df2, palette='viridis')
plt.xlabel('Rainfall Mean (Binned)')
plt.ylabel('PM2.5 Mean')
plt.title('Conditional Box Plot')
plt.show()
```
